```python
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
gru_df = pd.read_csv("mpp_gru_acc_heatmap.csv", index_col=0)
maxpool_df = pd.read_csv("mpp_maxpool_acc_heatmap.csv", index_col=0)
```

```python
# This setting works if you run this notebook in the environment with `Times New Roman` font.

sns.set(font="Times New Roman")
```

```python
fig = plt.figure(dpi=1200)
ax = fig.add_subplot(1, 1, 1)
sns.heatmap(gru_df, cmap="Blues", 
            annot=True, fmt=".3f",
            linewidths=.5, 
            square=True,
            vmin=0.0, vmax=1.0, 
            ax=ax,
            annot_kws={"size": 12})
ax.set(xlabel ='Ground Truth Missing Position',ylabel='Predicted Missing Position')

ax.xaxis.get_label().set_fontsize(14)
ax.yaxis.get_label().set_fontsize(14)
ax.tick_params(labelsize=12)

plt.savefig('mpp_heatmap_gru_no_title.eps', bbox_inches = 'tight', pad_inches=0.1)
plt.show()
```

```python
fig = plt.figure(dpi=1200)
ax = fig.add_subplot(1, 1, 1)
sns.heatmap(maxpool_df, cmap="Blues", 
            annot=True, fmt=".3f",
            linewidths=.5, 
            square=True,
            vmin=0.0, vmax=1.0, 
            ax=ax,
            annot_kws={"size": 12})
ax.set(xlabel ='Ground Truth Missing Position',ylabel='Predicted Missing Position')

ax.xaxis.get_label().set_fontsize(14)
ax.yaxis.get_label().set_fontsize(14)
ax.tick_params(labelsize=12)

plt.savefig('mpp_heatmap_maxpool_no_title.eps', bbox_inches = 'tight', pad_inches=0.1)
plt.show()
```
